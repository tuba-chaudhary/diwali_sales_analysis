## Average spending by age group

## "This chart highlights the age group with the highest average spending, providing valuable insights into consumer behavior. Such data is instrumental in identifying and understanding the target demographic more accurately, enabling more effective marketing and product positioning strategies."

```python
x = df.groupby("Age Group")["Amount"].mean()

# Plotting
plt.plot(x.index, x.values, color="blue", marker="o")
plt.xlabel("Age Group")
plt.ylabel("Average Amount Spent")
plt.title("Average Spending by Age Group")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gender-Based Spending Trends Across Age Demographics
## This chart highlights how customer spending varies across different age groups and genders, providing valuable insights into demographic purchasing behavior.

```python
Age_Group = df["Age Group"]
Gender = df["Gender"]
Amount = df["Amount"]
df2 = pd.DataFrame({
    "Age Group": Age_Group,
    "Gender": Gender,
    "Amount": Amount
})
df2
```

```python
female_df = df2[df2["Gender"] == "F"]
x = female_df.groupby("Age Group")["Amount"].mean()
male_df = df2[df2["Gender"] == "M"]
y = male_df.groupby("Age Group")["Amount"].mean()
plt.plot(x.index,x.values , color = "pink",label = "Female")
plt.plot(y.index,y.values,color = "blue",label = "Male")
plt.xlabel("Age Group")
plt.ylabel("Amount")
plt.title("Gender-Based Spending Trends Across Age Demographics")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Regional Spendings
## The pie chart illustrates regional spending patterns, highlighting how total customer expenditure is distributed across different geographic regions

```python
Region_spends = df.groupby("Region")["Amount"].mean().astype(int)
colour = ["#2F8783","#5948AB","#75216B","#C93A6D","#77C777"]
labels = Region_spends.index
```

```python
plt.pie(Region_spends.values,labels=  labels,colors = colour,autopct= "%1.1f%%")
plt.legend()
```

## Average Spending by Product Category
## This chart provides a clear visual comparison of product categories based on total spending, helping us identify which types of products are most popular among customers.

```python
categories = df.groupby("Product_Category")["Amount"].mean()
```

```python
plt.barh(categories.index,categories.values,label = "Category_spendings")
plt.xlabel("Categories of products",color = "purple")
plt.ylabel("Average spendings",color = "purple")
plt.title("Average Spending by Product Category",color = "red")
plt.tight_layout()
```
